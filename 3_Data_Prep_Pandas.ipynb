# Index-based selection :

* We use iloc to select data based on their numerical position in the dataframe.
* iloc takes two arguments, first is row followed by column. It has a starting index of 0 that is 0 is first, 1 is second, 2 is third and so on

```python
# First row and all columns
data.iloc[0,:3]
```

```python
# Fourth column that is the Name column and all rows
# Since starting index is 0 fourth column corresponds to index nnumber 3
data.iloc[:,3]
```

```python
# Second and third rows of the first
data.iloc[1:3,0]
```

```python
# First three rows and all columns
data.iloc[[0,1,2],:]
```

```python
# Bottom five rows of the dataframe
data.iloc[-5:,:]
```

# Label-based selection :

```python
# First row of the Name column
data.loc[0,'Name']
```

```python
# First 10 rows of the Name and Age column
data.loc[:9,['Name','Age']]
```

```python
# Rows with age > 40
data.loc[data['Age']>40,:]
```

```python
# Rows with age > 40 and are female
data.loc[(data['Age']>40)&(data['Sex']=='female'),:]
```

```python
# Rows with age > 40 OR have fare greater than or euql to 200
data.loc[(data['Age']>40)|(data['Fare']>=200),:]
```

```python
# All the rows with null cabin column
data.loc[data['Cabin'].isnull(),:]
```

```python
# All rows with C or Q in Embarked column
data.loc[data['Embarked'].isin(['C','Q']),:]
```

```python
# This is the same as if we had used the or statement
data.loc[(data['Embarked']=='C')|(data['Embarked']=='Q'),:]
```

# Handling Missing Data :

* We need to treat the missing data below Machine Learning modeling. If none of the missing value records are small, say less than 5%, then we can remove the records with missing value data by using the function pandas dropna().

```python
# Importing Pandas library

import pandas as pd
```

```python
# importing Titanic data

data = pd.read_csv("Titanic-Dataset.csv")
data.head()
```

```python
# Finding missing values

data.isnull().sum()
```

```python
# We see age and Cabin have missing values. As age is continuous 
# We can replace it with mean of the Age column
# Embarked and Cabin are categorical variables, we can use mode to replace missing values.

data.Age.fillna(data.Age.mean(),inplace=True)
data.Cabin.fillna(data.Embarked.mode()[0],inplace=True)
data.Embarked.fillna(data.Embarked.mode()[0],inplace=True)
data.isnull().sum()
```
